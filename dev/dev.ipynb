```python
from pydantic_agent import Agent, AgentSettings
from pydantic_agent.tools import glob_search, grep_search, run_bash
from pydantic_agent.tools.filesystem import (
    append_file,
    copy_file,
    delete_file,
    file_info,
    list_directory,
    move_file,
    read_file,
    write_file,
)

tools = [
    read_file,
    write_file,
    append_file,
    list_directory,
    file_info,
    delete_file,
    move_file,
    copy_file,
    run_bash,
    glob_search,
    grep_search,
]

settings = AgentSettings()

# agent = Agent.from_settings(settings,
#                             tools=tools)

agent = Agent(tools=tools,
              settings=settings)


result = await agent.run("What files are in the current directory?")

print(result.output)


```
