```python
from mamba_agents import Agent, AgentSettings
from mamba_agents.tools import glob_search, grep_search, run_bash
from mamba_agents.tools.filesystem import (
    append_file,
    copy_file,
    delete_file,
    file_info,
    list_directory,
    move_file,
    read_file,
    write_file,
)

tools = [
    read_file,
    write_file,
    append_file,
    list_directory,
    file_info,
    delete_file,
    move_file,
    copy_file,
    run_bash,
    glob_search,
    grep_search,
]

settings = AgentSettings()

agent = Agent(settings=settings,
              tools=tools)

result = await agent.run("What files are in the current directory?")

print(result.output)
```
